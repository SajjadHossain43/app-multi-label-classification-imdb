```python
all_genre = [genre for genres in df.genres for genre in eval(genres)]
```

```python
genre_counts = pd.Series(all_genre).value_counts().to_dict()
len(genre_counts)
```

```python
threshold = int(df.shape[0] * 0.01)
threshold
```

```python
genre_counts
```

```python
genres = df.genres.to_list()
revised_genres = []
for genres in genres:
  rev_genres = []
  genres = eval(genres)
  for genre in genres:
    if genre_counts[genre] >= threshold:
      rev_genres.append(genre)
  revised_genres.append(rev_genres)
revised_genres
```

```python
any(not genres for genres in revised_genres)
```

```python
sum(not genres for genres in revised_genres)
```

```python
df['revised_genres'] = revised_genres
```

```python
indices_to_drop = df[df.revised_genres.apply(lambda x: len(x) == 0)].index
type(indices_to_drop)
```

```python
x = df.drop(indices_to_drop).reset_index(drop=True)
x.shape, df.shape
```

```python
df = x
df
```
